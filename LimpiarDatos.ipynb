```python
#Meter la hoja DDD como lista de diccionarios
def DDD_to_Diccionario(Diccionario_ambiental):
    separador=";"
    #archivo = pd.read_csv('
    with open(Diccionario_ambiental, encoding="utf-8") as archivo:
        next(archivo) #Omito el encabezado
        Datos = []
        for linea in archivo:
            linea = linea.rstrip("\n")
            columnas = linea.split(separador)
            Nombre = columnas[0]
            Etiqueta=str(columnas[1])
            Medida=str(columnas[2])
            Datos.append({
                "Nombre": Nombre,
                "Etiqueta": Etiqueta,
                "Medida": Medida, 
            })
        return Datos

def main():
    Datos = DDD_to_Diccionario("C:\BD_Diccionario.csv")
    for Datos in Datos:
        Nombre = Datos["Nombre"]
        Etiqueta = Datos["Etiqueta"]
        Medida = Datos["Medida"]
        print (Datos)
main()
```
